```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

R=47
```

```python
def func(f,C,theta0):
    return np.arctan(2*np.pi*R*C*f) + theta0
```

```python
f=np.arange(1,10.2,.2);f=f*1000
```

```python
theta=np.array([-71.42,
-72.1,
-72.3,
-72.22,
-71.9,
-71.51,
-71.01,
-70.45,
-69.81,
-69.16,
-68.46,
-67.78,
-67.05,
-66.31,
-65.58,
-64.84,
-63.9,
-63.25,
-62.54,
-61.91,
-61.23,
-60.51,
-59.79,
-58.92,
-58.34,
-57.66,
-56.99,
-56.38,
-55.73,
-55.09,
-54.41,
-53.79,
-53.22,
-52.71,
-52.18,
-51.52,
-51.02,
-50.51,
-49.98,
-49.41,
-48.9,
-48.41,
-47.84,
-47.33,
-46.84,
-46.34])

theta=theta*np.pi/180
```

```python
[popt,pcov] = curve_fit(func,f,theta,p0=[25e-12,-np.pi/2])
popt,pcov
```

```python
plt.scatter(f,theta,marker='x',label='experiment')
plt.plot(f,func(f,*popt),color='red',linewidth=2,label='curve fitting')
plt.xlabel('f (kHz)'); plt.ylabel('$\theta$')
plt.title('Graph of reference angle and frequency')
plt.grid()
plt.legend(loc='best')
```
